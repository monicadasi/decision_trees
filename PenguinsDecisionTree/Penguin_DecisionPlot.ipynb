```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Load the dataset
df_penguin = pd.read_csv("penguins_size.csv")
df_penguin = df_penguin.dropna()

df_penguin = df_penguin[df_penguin["gender"] != '.']
df_penguin['gender'].unique()

# The reason for using two-dimensional numerical data is that we want to show the examples visually.
df_penguin_X = df_penguin[['culmen_length_mm','culmen_depth_mm']]
df_penguin_y = df_penguin['gender']

# convert dataframe to numpy array
df_penguin_X = df_penguin_X.to_numpy()
df_penguin_y = df_penguin_y.astype('category').cat.codes
# df_penguin_X
```

```python
def plot_decision_boundary(clf, X, Y, cmap='Paired_r'):
    h = 0.02
    x_min, x_max = X[:,0].min() - 10*h, X[:,0].max() + 10*h
    y_min, y_max = X[:,1].min() - 10*h, X[:,1].max() + 10*h
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)


    plt.figure(figsize=(8,6))
    plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.25)
    plt.contour(xx, yy, Z, colors='k', linewidths=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=cmap, edgecolors='k')   
#     sns.scatterplot(data=df_penguin, x="culmen_length_mm", y="culmen_depth_mm", hue="gender")
    
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    
    plt.show()
```

### Decision Surface of Decision Tree trained on pair of features

```python

# Fit on classifier
tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(df_penguin_X, df_penguin_y)

plot_decision_boundary(tree_clf, df_penguin_X, df_penguin_y)

plot_decision_boundary1(tree_clf, df_penguin_X, df_penguin_y)
```
